```python
import numpy as np
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from bct import modularity, clustering, centrality
import statsmodels.stats.api as sms
from sklearn.linear_model import LinearRegression
from statsmodels.stats.multitest import fdrcorrection
```

```python
demo = pd.read_csv("../dataset/Study3/List_sub_data_1T.csv")
demo2 = pd.read_csv("../dataset/Study3/List_sub_data_1T_vXCP-D_v2.csv")
```

```python
demo = pd.merge(demo, demo2.loc[:, ['Code', 'SES']], on='Code', how='left')
```

```python
motion = pd.read_csv('../dataset/Study3/MotionParameter.csv')
motion = motion.iloc[:, :6].dropna()
motion['ID'] = [x.replace('Sub', 'sub') for x in motion.Participants.values]
motion['meanFD'] = [float(x) for x in motion['Unnamed: 3']]
motion['signal_quality'] = motion['Unnamed: 5']
```

```python
fc = glob.glob('../dataset/Study3/correlation/sub-*_ses-*_task-rs_run-*_space-MNI152NLin2009cAsym_seg-4S456Parcels_stat-pearsoncorrelation_relmat.tsv')
```

```python
corall = []
ids = []
for f in fc:
    df = pd.read_csv(f, sep='\t')
    df = df.iloc[:400, 1:401]
    df.columns = [x.split('_')[1] for x in df.columns.values]
    df.index = df.columns.values
    df_long = df.reset_index().melt(id_vars='index', var_name='column', value_name='value')
    avg_network_corr = df_long.groupby(['index', 'column'])['value'].mean().values
    corall.append(avg_network_corr)
    ids.append(f.split('/')[-2])
```

```python
network_names = ['Cont', 'Default',  'DorsAttn', 'Limbic','SalVentAttn', 'SomMot', 'Visual']
labels = [f"{i}_{j}" for i in network_names for j in network_names]
df_corall = pd.DataFrame(corall, columns=labels)
df_corall['ID'] = ids
```

```python
yeo = df_corall.copy()
```

```python
demo = demo.dropna(subset=['list_sub.txt'])
demo['ID'] = [x.split('_')[0].replace('-ses-2', '') for x in demo['list_sub.txt']]
```

```python
yeo = pd.merge(yeo, demo, on='ID')
yeo = pd.merge(yeo, motion, on='ID')
```

```python
#yeo = yeo[yeo['meanFD']<1]
yeo = yeo[yeo['signal_quality']=='good']
yeo = yeo[yeo['Age']<12]
yeo['fluid_intelligence'] = pd.to_numeric(yeo['fluid_intelligence'], errors='coerce')
yeo['playing_outside'] = pd.to_numeric(yeo['playing_outside'], errors='coerce')
```

```python
yeo.to_csv('../dataset/Study3/study3_all_yeo7.csv', index=False)
```

```python
# Define the networks of interest
networks = ['Default_Default', 'Default_SalVentAttn', 'Default_Cont', 
            'Cont_Cont', 'Cont_SalVentAttn', 'SalVentAttn_SalVentAttn']

# Convert Pedagogy to numeric
yeo['Pedagogy'] = yeo['Pedagogy'].map({'M': 1, 'T': 0})

# Store results for heatmaps
all_results = {}

for target in ['Pedagogy', 'SES']: #'fluid_intelligence', 'playing_outside', 'fluid_intelligence', 
    # Setup
    X = yeo[networks].copy()
    y = pd.to_numeric(yeo[target], errors='coerce')

    # Covariates
    if target != 'Pedagogy':
        covariates_X = yeo[['Age', 'Gender', 'meanFD', 'Pedagogy', 'SES']].copy() #'SES', 
        covariates_y = yeo[['Age', 'Gender', 'Pedagogy']].copy()
    else:
        covariates_X = yeo[['Age', 'Gender', 'meanFD', 'SES']].copy() #, 'fluid_intelligence'
        covariates_y = yeo[['Age', 'Gender', 'SES']].copy()
    
    covariates_X['Gender'] = covariates_X['Gender'].map({'F': 1, 'M': 0})
    covariates_y['Gender'] = covariates_y['Gender'].map({'F': 1, 'M': 0})

    # Drop missing data
    data = pd.concat([X, y, covariates_X], axis=1).dropna()
    X = data[networks]
    y = data[y.name]
    if target != 'Pedagogy':
        covariates_X = data[['Age', 'Gender', 'meanFD',  'Pedagogy']]
        covariates_y = data[['Age', 'Gender', 'Pedagogy']]
    else:
        covariates_X = data[['Age', 'Gender',  'meanFD', 'SES']]
        covariates_y = data[['Age', 'Gender', 'SES']]

    # Residualize X
    X_resid = pd.DataFrame(index=X.index)
    for col in X.columns:
        model = LinearRegression().fit(covariates_X, X[col])
        X_resid[col] = X[col] - model.predict(covariates_X)

    # Residualize y
    y_model = LinearRegression().fit(covariates_y, y)
    y_resid = y - y_model.predict(covariates_y)

    # Run regression for each network
    results = []
    for col in X_resid.columns:
        X_single = sm.add_constant(X_resid[col])
        model = sm.OLS(y_resid, X_single).fit()
        coef = model.params[col]
        pval = model.pvalues[col]
        results.append({'network': col, 'coef': coef, 'pval': pval})

    results_df = pd.DataFrame(results)
    rejected, pvals_corrected = fdrcorrection(results_df['pval'])
    results_df['pval_fdr'] = pvals_corrected
    results_df['significant'] = rejected

    # Store for heatmap
    all_results[target] = results_df.set_index('network')[['coef', 'pval_fdr', 'significant']]

    # Print results
    print(f"\n--- {target} ---")
    print(results_df.sort_values('pval_fdr'))

# Create DataFrame of coefficients for each target
coef_matrix = pd.DataFrame({key: df['coef'] for key, df in all_results.items()})

# Plot heatmap (all values shown, no masking)
plt.figure(figsize=(8, 4))
sns.heatmap(coef_matrix.T, annot=True, center=0, cmap='coolwarm', fmt=".2f",
            cbar_kws={'label': 'Beta Coefficient'})
plt.title("Regression Coefficients by Target and Network")
plt.ylabel("Target Variable")
plt.xlabel("Network")
plt.tight_layout()
plt.savefig('heatmap.eps')
```

```python
from scipy.stats import ttest_ind

for network in ['Default_SalVentAttn', 'Default_Cont', 'SES']:
    # make sure data is numeric
    yeo[network] = yeo[network].astype(float)
    
    # split by Pedagogy group
    group0 = yeo.loc[yeo['Pedagogy'] == 0, network].dropna()
    group1 = yeo.loc[yeo['Pedagogy'] == 1, network].dropna()
    
    # t-test
    t, p = ttest_ind(group0, group1, equal_var=False)
    print(f"{network}: t = {t:.3f}, p = {p:.4f}, n0 = {len(group0)}, n1 = {len(group1)}")
    
    # plot
    figure = plt.figure(figsize=(4,5))
    sns.boxplot(x='Pedagogy', y=network, data=yeo, color='lightgray')
    sns.stripplot(x='Pedagogy', y=network, data=yeo,
                  color='blue', size=6, jitter=True)
    plt.xlabel("Pedagogy (0=T, 1=M)")
    plt.ylabel(network)
    plt.tight_layout()
    plt.savefig(network+'.eps')
    plt.close()
```

```python
C_all = []
P_all = []
ids = []
for t, cor in enumerate(fc):
    if t % 20 == 0:
        print(t / len(fc))
    df = pd.read_csv(cor, sep='\t').iloc[:400, 1:401].fillna(0)
    df.columns = np.arange(400)
    # Identify valid rows/columns
    valid_rows = ~(df == 0).all(axis=1)
    valid_cols = ~(df == 0).all(axis=0)
    df_valid = df.loc[valid_rows, df.columns[valid_cols]]
    if df_valid.shape[0] < 2 or df_valid.shape[0] != df_valid.shape[1]:
        print(f"Skipping (non-square or too small): {cor}")
        C_all.append([None] * 400)
        P_all.append([None] * 400)
        continue
    df_matrix = np.tanh(df_valid.to_numpy())
    ci, q = modularity.community_louvain(df_matrix, B='negative_asym')
    clustering_vals = np.nanmean(clustering.clustering_coef_wu_sign(df_matrix), axis=0)
    participation_vals = np.nanmean(centrality.participation_coef_sign(df_matrix, ci), axis=0)
    # Fill full 400-length vectors
    C_full = [None] * 400
    P_full = [None] * 400
    valid_idx = valid_rows[valid_rows].index.tolist()
    for i, idx in enumerate(valid_idx):
        C_full[idx] = clustering_vals[i]
        P_full[idx] = participation_vals[i]
    C_all.append(C_full)
    P_all.append(P_full)
    ids.append(cor.split('/')[-2])
```

```python
df_c_human = pd.DataFrame(C_all)
df_c_human.columns = ['clust'+str(x) for x in np.arange(1,401)]
df_c_human['ID'] = ids
df_p_human = pd.DataFrame(P_all)
df_p_human.columns = ['part'+str(x) for x in np.arange(1,401)]
df_p_human['ID'] = ids
df_c_p_human = pd.merge(df_c_human, df_p_human, on='ID')
```

```python
df_c_p_human = pd.merge(df_c_p_human, demo, on='ID')
df_c_p_human = pd.merge(df_c_p_human, motion, on='ID')
df_c_p_human = df_c_p_human[df_c_p_human['signal_quality']=='good']
df_c_p_human = df_c_p_human[df_c_p_human.Age<12]
```
