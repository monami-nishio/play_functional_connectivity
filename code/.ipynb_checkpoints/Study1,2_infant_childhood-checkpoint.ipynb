```python
def map_component(x):
    if 1 <= x <= 2:
        return 1
    elif 3 <= x <= 4:
        return 2
    elif 5 <= x <= 6:
        return 3
    elif 7 <= x <= 8:
        return 4
    elif 9 <= x <= 10:
        return 5
    elif 11 <= x <= 13:
        return 6
    elif 14 <= x <= 17:
        return 7
    else:
        return np.nan
region_label['schaefer17_combined'] = [map_component(x) for x in region_label.schaefer17]
```

```python
# clustering, participation
C_all = []
P_all = []
ids = []
sess = []
for t, cor in enumerate(CBPD):
    if t % 20 == 0:
        print(t / len(CBPD))
    try:
        df = pd.read_csv(cor, sep='\t').iloc[:400, 1:401].fillna(0)
        df.columns = np.arange(400)
        # Identify valid rows/columns
        valid_rows = ~(df == 0).all(axis=1)
        valid_cols = ~(df == 0).all(axis=0)
        df_valid = df.loc[valid_rows, df.columns[valid_cols]]
        if df_valid.shape[0] < 2 or df_valid.shape[0] != df_valid.shape[1]:
            print(f"Skipping (non-square or too small): {cor}")
            C_all.append([None] * 400)
            P_all.append([None] * 400)
            continue
        df_matrix = np.tanh(df_valid.to_numpy())
        try:
            ci, q = modularity.community_louvain(df_matrix, B='negative_asym')
            clustering_vals = np.mean(clustering.clustering_coef_wu_sign(df_matrix), axis=0)
            participation_vals = np.mean(centrality.participation_coef_sign(df_matrix, ci), axis=0)
        except Exception as e:
            print(f"Skipping (error in computation): {cor}, Error: {e}")
            C_all.append([None] * 400)
            P_all.append([None] * 400)
            continue
        # Fill full 400-length vectors
        C_full = [None] * 400
        P_full = [None] * 400
        valid_idx = valid_rows[valid_rows].index.tolist()
        for i, idx in enumerate(valid_idx):
            C_full[idx] = clustering_vals[i]
            P_full[idx] = participation_vals[i]
        C_all.append(C_full)
        P_all.append(P_full)
        ids.append(os.path.basename(cor))
    except:
        continue
```

```python
def compute_slope(group, region):
    return np.polyfit(group['scan_age'], group[region], 1)[0]  # slope
slope_df = []
for region in ['Default_Default',
     'Default_Cont',
     'Default_DorsAttn',
     'Default_SalVentAttn',
     'Default_Limbic',
     'Default_SomMot',
     'Default_Visual']:
    slopes = (
        bcp_all[bcp_all.interview_age<600][bcp_all.scan_age<600].groupby('sub_id')
        .apply(lambda g: compute_slope(g, region))
        .rename(region + '_slope')
    )
    slope_df.append(slopes)
slope_df = pd.concat(slope_df, axis=1)
```

```python
def compute_slope(group, region):
    x = pd.to_numeric(group['interview_age'], errors='coerce')
    y = pd.to_numeric(group[region], errors='coerce')
    valid = ~(x.isna() | y.isna())
    if valid.sum() < 2:
        return np.nan  # Not enough points to compute slope
    return np.polyfit(x[valid], y[valid], 1)[0]
slope_word_df = []
for region in ['imitation_play_percentile']:
    slopes = (
        bcp_all[bcp_all.interview_age<600][bcp_all.scan_age<600].dropna(subset=[region]).groupby('src_subject_id')
        .apply(lambda g: compute_slope(g, region))
        .rename(region + '_slope')
    )
    slope_word_df.append(slopes)
slope_word_df = pd.concat(slope_word_df, axis=1)
```

```python
avg_across = (
    bcp_all
    .groupby('src_subject_id')[['interview_age']]
    .mean()
    .rename(columns=lambda x: 'avg_' + x)
)
```

```python
slope_word_df = slope_word_df.join(avg_across).reset_index()
```
